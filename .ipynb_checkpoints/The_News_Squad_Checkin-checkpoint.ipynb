Load Data
-----

```python
# general imports
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
pd.set_option('display.max_columns', 62)
```

```python
# load data
news = pd.read_csv("OnlineNewsPopularity.csv")
news.columns = [col.strip(" ") for col in news.columns]  # delete the white space before each column name
news.head()
```

```python
# choose X, y
X = news.iloc[:, 1:60]  # all columns except url and target
y = news["shares"].values.ravel()  # target: number of shares
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
```

Fit scikit-learn model
----

```python
# choose a random forest regressor
model = RandomForestRegressor(n_estimators=100, min_samples_leaf=50)
model.fit(X_train, y_train)
```

Evaluation Metric
----

```python
# predict y values on test data
y_pred = model.predict(X_test)
```

```python
# print out the R^2 of the prediction
model.score(X_test, y_test)
```

```python
# choose median absolute error as the evaluation metric
mae = mean_absolute_error(y_test, y_pred)
mae
```
